# Simple plots to visualise audio input data

```python
import numpy as np
%matplotlib inline
import pylab as plt

#
#load audio data
#
data = np.loadtxt('../samples/sample_sines.txt')
samples = data[:,None]
print(samples)
```

```python
#
# load FFT data amplitudes
#
data = np.loadtxt('../samples/sine_amplitudes.txt')
ffta = []
n_frames = len(data)/513

for i in range(int(n_frames)):
    ffta.append(data[i*513:(i+1)*513])
print(ffta[977])
```

```python
#
# load FFT data phases
#
data = np.loadtxt('../samples/sine_phases.txt')
fftp = []
n_frames = len(data)/513

for i in range(int(n_frames)):
    fftp.append(data[i*513:(i+1)*513])
print(fftp[977])
```

```python
frame  = fftp[977][:,None]
frame[0] = 0
plt.plot(frame)
```

```python
frame  = ffta[977][:,None]
frame[0] = 0
plt.plot(frame)
```

```python
result = np.where(frame == frame.min())
print(result)
```
