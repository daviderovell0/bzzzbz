# Simple plots to visualise audio input data

```python
import numpy as np
%matplotlib inline
import pylab as plt

#
#load audio data
#
data = np.loadtxt('../samples/sample_sines.txt')
samples = data[:,None]
print(samples)
```

```python
#
# load FFT data amplitudes
#
data = np.loadtxt('../samples/sine_amplitudes.txt')
ffta = []
n_frames = len(data)/513

for i in range(int(n_frames)):
    ffta.append(data[i*513:(i+1)*513])
print(ffta[977])
```

```python
#
# load FFT data phases
#
data = np.loadtxt('../samples/sine_magnitudes.txt')
fftm = []
n_frames = len(data)/513

for i in range(int(n_frames)):
    fftm.append(data[i*513:(i+1)*513])
print(fftm[977])
```

```python
frame  = fftm[977][:,None]
frame[0] = 0
plt.yscale("log")
plt.plot(frame)
```

```python
frame  = fftm[391][:,None]
frame[0] = 0
#plt.yscale("log")
plt.plot(frame[11:])
```

```python
result = np.where(frame == frame.min())
print(result)
```
